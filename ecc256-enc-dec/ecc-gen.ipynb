```python
#Generate Checksum
for i in range(H.shape[0]):
    print(f"CHK[{i}] = ",end="")
    for j in range(H.shape[1]):
        if H[i,j]:
            print(f"IN[{j}] ^ ",end="")
    print(" 1'b0 ;")
```

```python
#Generate Syndrome
for i in range(H.shape[0]):
    print(f"SYN[{i}] = ",end="")
    for j in range(H.shape[1]):
        if H[i,j]:
            print(f"IN[{j}] ^ ",end="")
    print(f"CHK[{i}];")
```

```python
def get_random_vector(size):
    return np.random.randint(0,2,size,dtype=np.int8)

def compute_chksum(a,H):
    chksum = np.zeros(H.shape[0],dtype=np.int8)
    for i in range(0,H.shape[0]):
        for j in range(0,H.shape[1]):
            if H[i,j]:
                chksum[i] = chksum[i] ^ a[j]
    return chksum

def compute_syndrome_given_checksum(a,chk,H):
    syn = np.zeros(H.shape[0],dtype=np.int8)
    for i in range(0,H.shape[0]):
        syn[i] = chk[i]
        for j in range(0,H.shape[1]):
            if H[i,j]:
                syn[i] = syn[i] ^ a[j]
    return syn

def compute_syndrome(enc,H):
    a = enc[0:H.shape[1]]
    chk = enc[H.shape[1]:]
    return compute_syndrome_given_checksum(a,chk,H)

def flip(a,posn):
    if a[posn]:
        a[posn] = 0
    else:
        a[posn] = 1
    return a

def concat(a,chk):
    return np.concatenate((a,chk))

def correction_hash():
    chash = {}
    for posn in range(0,272):
        a = get_random_vector(256)
        chk = compute_chksum(a,H)
        dec_in = concat(a,chk)
        dec_err = flip(dec_in,posn)
        syn = compute_syndrome(dec_err,H)
        key = ''.join(map(str,syn))
        chash[key] = posn
    return chash

def correct(enc,chash):
    syn = compute_syndrome(enc,H)
    key = ''.join(map(str,syn))
    posn = chash[key]
    return flip(enc,posn)
```

```python
chash = correction_hash()
```

```python
for posn in range(1,272):
    a = get_random_vector(256)
    chk = compute_chksum(a,H)
    enc = concat(a,chk)
    enc = flip(enc,posn)
    dec_out = correct(enc,chash)
    check_correction = np.all(dec_out == concat(a,chk))
    assert check_correction , f"flip {posn}: {check_correction}"
```
